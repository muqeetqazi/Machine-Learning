```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
canada_price=pd.read_csv("canadahomeprice.csv")
canada_price.columns
canada_price = canada_price.rename(columns={'Unnamed: 1': 'price'})
canada_price
```

```python
new_area=canada_price.drop('price',axis='columns')
new_price=canada_price.drop('year',axis='columns')
```

```python
model=LinearRegression()
model.fit(new_area,new_price)
```

```python
new_data = pd.DataFrame({'year': [2050]})
predicted_price = model.predict(new_data)
print(predicted_price)
```

```python
plt.xlabel('year')
plt.ylabel('price')
plt.scatter(canada_price['year'], canada_price['price'], color='red', marker='+')
plt.plot(canada_price['year'], model.predict(canada_price[['year']]), color='blue') 
plt.show()
```
